```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('Titanic.csv',usecols=['Survived','Cabin'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
df.head(10)
```

```python
df['Cabin'].unique()
```

```python
#The code `df['Cabin']=df['Cabin'].astype(str).str[0]` is converting the 'Cabin' column in the DataFrame 'df' to a string data type and then extracting the first character of each string.
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df['Cabin'].unique()
```

```python
prob_df = df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df = pd.DataFrame(prob_df)
prob_df
```

```python
#The code `prob_df['Died']=1-prob_df['Survived']` is creating a new column in the `prob_df` DataFrame called 'Died'. The values in this column are calculated by subtracting the values in the 'Survived' column from 1. This is done to represent the opposite outcome of survival, which is death. The resulting DataFrame is then displayed.
prob_df['Died']=1-prob_df['Survived']
prob_df
```

```python
#The code `prob_df['Prob Encoding']=prob_df['Survived']/prob_df['Died']` is creating a new column called 'Prob Encoding' in the DataFrame `prob_df`. The values in this new column are calculated by dividing the values in the 'Survived' column by the values in the 'Died' column.
prob_df['Prob Encoding']=prob_df['Survived']/prob_df['Died']
prob_df
```

```python
prob_encoding = prob_df['Prob Encoding'].to_dict()
```

```python
df['Cabin Encoded'] = df['Cabin'].map(prob_encoding)
df.head()
```
