##### Loading the dataset

```python
top_songs = pd.read_csv('spotify-2023.csv',encoding='latin-1')
top_songs.head()
```

##### Filling null values with stats data

```python
# Filling null values in Shazam charts column
top_songs['in_shazam_charts'] = pd.to_numeric(top_songs['in_shazam_charts'],errors='coerce')
top_songs['in_shazam_charts']=top_songs['in_shazam_charts'].fillna(int('51'))
top_songs['in_shazam_charts']= top_songs['in_shazam_charts'].astype('int64')
top_songs['in_shazam_charts'].isnull().sum()
```

```python
# Filling null values in Streams column
top_songs['streams'] = pd.to_numeric(top_songs['streams'],errors='coerce')
top_songs['streams']=top_songs['streams'].fillna(int('514137424'))
top_songs['streams']=top_songs['streams'].astype('int64')
```

```python
# Filling null values in Deezer charts column
top_songs['in_deezer_playlists'] = pd.to_numeric(top_songs['in_deezer_playlists'],errors='coerce')
top_songs['in_deezer_playlists']=top_songs['in_deezer_playlists'].fillna(109)
top_songs['in_deezer_playlists']=top_songs['in_deezer_playlists'].astype('Int64')
```

### Top 10 songs in spotify by presence in spotify playlists

```python
# Top ten songs in spotify playlists
top_in_spotify = top_songs.sort_values(by='in_spotify_playlists',ascending=False).head(10)
plt.figure(figsize=(10,4))
sns.barplot(x=top_in_spotify['in_spotify_playlists'].values,y=top_in_spotify['track_name'],palette='mako')
plt.title('Top 10 songs in spotify playlists')
plt.xlabel('Playlists Count')
plt.ylabel('Songs Name')
plt.tight_layout()
```

#### Observations
- Get Lucky -Radio Edit has the highest presence in spotify playlists by being in more than 55000+ playlists.
- On No.2 Mr.Brightside has the second highest presence with 50000+ playlists.
- On no.3 Wake me up -Radio edit is the third song with the most presence in spotify playlists

#### Analyzing increase in number of songs released every year

```python
releases_by_year = top_songs['released_year'].value_counts()
plt.figure(figsize=(6,4))
sns.lineplot(x=releases_by_year.index,y=releases_by_year.values)
plt.title('Number of Songs released by year')
plt.xlabel('Year')
plt.ylabel('Number of songs released')
plt.tight_layout()
plt.show()
```

#### Observation 
By analyzing this line chart , we can say that the number of songs released every year has increased over the years.
Around 1940, number of songs releases were around 1-4 and as years kept passing by, in 2020, number of songs released every year has drastically increased to over 350+ songs released every year.

```python
# Using list comprehension to find columns with certain strings in it
songs_features = [col for col in top_songs.columns if '%' in col]
songs_charts = [col for col in top_songs.columns if 'in_' in col]
```

### Percentage features Distribution analysis

```python
for feature in songs_features:
    plt.figure(figsize=(8,3))
    sns.histplot(top_songs[feature],bins=30,kde=True)
    plt.title(f'Distribution of {feature}',fontsize=12)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

### Top artists based on their streams

```python

top_artists_by_streams = top_songs.sort_values(by='streams',ascending=False).head(10)
top_artists_by_streams['artist(s)_name']

plt.figure(figsize=(6,4))
sns.barplot(x=top_artists_by_streams['streams'].values,y=top_artists_by_streams['artist(s)_name'],palette='plasma')
plt.xlabel('Total Streams (in billions) ')
plt.ylabel("Artist's name")
plt.show()
```

#### Features correlation with Streams

```python
for feature in songs_features:
    plt.figure(figsize=(8,3))
    sns.scatterplot(x=top_songs[feature],y=top_songs['streams'],alpha=0.6)
    plt.title(f'Streams vs {feature}',fontsize=12)
    plt.xlabel(feature)
    plt.ylabel('Streams')
    corr = top_songs['streams'].corr(top_songs[feature])
    plt.annotate(f'Correlation :{corr:.2f}',xy=(0.05,0.9),xycoords = 'axes fraction',fontsize=12)
    plt.tight_layout()
    plt.show()
```

### Features Correlation with heatmap

```python
features_corr = top_songs[songs_features].corr()
sns.heatmap(features_corr,annot=True)
plt.show()
```

#### Analyzing Top 10 artists based on songs they have created and plotting them according to their songs features

```python
top_artists_by_songs = top_songs['artist(s)_name'].value_counts().sort_values(ascending=False).head(10).index
top_artist = top_songs[top_songs['artist(s)_name'].isin(top_artists_by_songs)]
for feature in songs_features:
    if feature != 'instrumentalness_%':
        feature_average = top_artist.groupby('artist(s)_name')[feature].mean().sort_values(ascending=False).values
        sns.barplot(x=feature_average,y=top_artists_by_songs,palette='flare')
        plt.xlabel(feature)
        plt.title(f'Top 10 Artists with their {feature} features')
        plt.show()
```

### Observation on artists by their songs percentage features
- Taylor Swift seems to have the highest percentage in all of the percentage features
- Coming on number 2 , The Weeknd seems to be the second highest in percentage features
- And at last Drake, 21 Savage seems to have the lowest score in percentage feature in his songs
