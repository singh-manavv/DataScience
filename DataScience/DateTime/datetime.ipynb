```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import datetime
```

```python
today_date = datetime.datetime.today()
```

```python
today_date
```

```python
today_date-datetime.timedelta(3)
```

```python
days = [today_date-datetime.timedelta(x) for x in range(0,15)]
```

```python
days
```

```python
df = pd.DataFrame(days)
```

```python
df.columns = ['Days']
df
```

```python
df['WeekDay'] = df['Days'].dt.day_name()
df
```

```python
#The code is creating a dictionary called `dict` with the keys as the days of the week (Monday, Tuesday, etc.) and the values as their corresponding numerical representation (1, 2, etc.).
day_names = {
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5,
    'Saturday':6,
    'Sunday':7
    
}

day_names
```

```python
df['Weekday_ordinal'] = df['WeekDay'].map(day_names)
df
```

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set
```

```python
columns = [1,3,5,6,7,8,9,13]
train_set = train_set[columns]
train_set.columns = ['Employment','Degree','Status','Designation','family','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns:
    print(f'{feature} : {len(train_set[feature].unique())} labels')
```

```python
country_map = train_set['Country'].value_counts().to_dict()
country_map
```

```python
train_set['Country'] = train_set['Country'].map(country_map)
```
